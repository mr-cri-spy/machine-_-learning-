```python
### Handle Null Values
```

```python
data.isnull().sum()
```

```python
## Remove row/column having null value
```

```python
len(data)
```

```python
## remove all rows where you have null values
print(data.dropna(axis=0).shape)    ## axis = 0|1  row|column  -- default = 0
```

```python
## remove all colns where you have null values
print(data.dropna(axis=1).shape)    ## axis = 0|1  row|column
```

```python
## remove all rows where any column value is null
print(data.dropna(how="any").shape)    ## how = any | all  default = any
```

```python
## remove all rows where all column value is null (entire row is null)
print(data.dropna(how="all").shape)    ## how = any | all  default = any
```

```python
## remove all rows where Age column value is null
print(data.dropna(subset=["Age"]).shape)
```

```python
## remove all rows where Age column value is null
print(data.dropna(subset=["Embarked"]).shape)
```

```python
## remove all rows where Age column value is null
print(data.dropna(subset=["Age","Embarked"]).shape)
```

```python
## remove all rows where Age column value is null
print(data.dropna(subset=["Age","Embarked"],how="all").shape)
```

```python
mean_age = data['Age'].mean()
data['Age'] = data['Age'].fillna(mean_age)
```

```python
data['Cabin'] = data['Cabin'].fillna("NA")
```

```python
data['Embarked'].value_counts()
```

```python
data['Embarked'] = data['Embarked'].fillna("S")
```
