```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Histogram
```

```python
np.random.seed(40)
samples = np.random.randint(1,100,50)
print(samples)
plt.hist(samples)
plt.show()
```

```python
ax = sns.histplot(samples)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
### Pie Chart
```

```python
gender = tips['sex'].value_counts()
gender
```

```python
gender.values
```

```python
gender.index
```

```python
plt.pie(gender.values,labels = gender.index,autopct="%0.2f%%")
plt.show()
```

```python
days = tips['day'].value_counts()
days
```

```python
plt.pie(days.values,labels = days.index,autopct="%0.2f%%",explode=[0,0,0,0.2])
plt.show()
```

```python
sns.countplot(x=tips['sex'])
plt.show()
```

```python
sns.countplot(x=tips['sex'],hue=tips['smoker'])
plt.show()
```

```python
sns.countplot(x=tips['sex'],hue=tips['day'])
plt.show()
```

```python
### Line Plot
```

```python
x = [3,8,9,10,15]
y = [30,40,50,10,100]
plt.plot(x,y)
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],s=90,hue=tips['sex'])
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],s=90,hue=tips['day'])
plt.show()
```

```python
### Rel Plot
```

```python
sns.relplot(x=tips['total_bill'],y=tips['tip'],kind="line")
plt.show()
```

```python
sns.relplot(x=tips['total_bill'],y=tips['tip'],kind="scatter",s=150,hue=tips['day'])
plt.show()
```

```python
sns.relplot(x=tips['total_bill'],y=tips['tip'],kind="scatter",s=150,hue=tips['day'],col=tips['day'],col_wrap=4)
plt.show()
```

```python
sns.relplot(x=tips['total_bill'],y=tips['tip'],kind="scatter",s=150,hue=tips['day'],col=tips['day'],col_wrap=2)
plt.show()
```

```python
### Heatmap
np.random.seed(10)
samples = np.random.randint(1,100,(10,10))
samples
```

```python
sns.heatmap(samples,annot=True,cmap="Reds")
plt.show()
```

```python
sns.heatmap(samples,annot=True,cmap="Reds",vmin=50,vmax=100)
plt.show()
```
