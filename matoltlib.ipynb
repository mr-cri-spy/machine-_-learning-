```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
img = np.ones((512,512,3))* 0.5
cv2.rectangle(img,(100,400),(400,200),(255,0,0),-1)

plt.imshow(img)
plt.show()
```

```python
img = np.ones((512,512,3))* 0.5
cv2.rectangle(img,(100,400),(400,200),(255,0,0),-1)
cv2.line(img,(100,400),(400,200),(255,255,255),10)
plt.imshow(img)
plt.show()
```

```python
img = np.ones((512,512,3))* 0.5
cv2.circle(img,(255,255),200,(255,0,0),-1)
cv2.circle(img,(255,255),2,(255,255,255),3)
cv2.line(img,(255,255),(455,255),(255,255,255),2)
plt.imshow(img)
plt.show()
```

###Draw polygon

```python
img = np.ones((512,512,3),np.uint8)* 255
pts = [(250,100),(100,300),(200,300), (400,400)]
cv2.fillPoly(img,np.array([pts]),(255,0,0))
plt.imshow(img)
plt.show()
```
